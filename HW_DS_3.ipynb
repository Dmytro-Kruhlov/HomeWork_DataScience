```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Напишіть функцію гіпотези лінійної регресії у векторному вигляді;

```python
df = pd.read_csv("/content/Housing.csv")
print(df[df.isna()].count())
print(df.describe())
df
```

```python
x_col = ["area", "bedrooms","bathrooms"]
y_col = "price"
data = [y_col] + x_col
df_view = df[data]
df_view
```

```python
x_data = df[x_col].values
print(x_data.shape)
print(x_data)
y_data = df[y_col].values.reshape(-1, 1)
print(y_data.shape)
weight = np.zeros(3).reshape(-1, 1)
print(weight.reshape(-1, 1).shape)
learning_rate = 0.0000000001
num_iterations = 50

```

```python
def hypothesis(w, X):
  return np.dot(X, w)
```

```python
hyp = hypothesis(weight, x_data)
hyp.shape
```

Створіть функцію для обчислення функції втрат у векторному вигляді;

```python
def loss_fuction(w, X, y):
    m = len(y)
    predictions = hypothesis(w, X)
    cost = (1/(2*m)) * np.sum(np.square(predictions - y))
    return cost
```

```python
loss = loss_fuction(weight, x_data, y_data)
print(loss)
```

Реалізуйте один крок градієнтного спуску;

```python
def gradient_descent_step(w, X, y, learning_rate):
    m = len(y)
    predictions = hypothesis(w, X)
    errors = predictions - y
    gradient = (1/m) * np.dot(X.T, errors)

    w = w - (learning_rate * gradient)
    return w
```

```python
step = gradient_descent_step(weight, x_data, y_data, learning_rate)
print(step)
```

Знайдіть найкращі параметри $\vec{w}$ для датасету прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень;

```python
def gradient_descent(X, y, w, learning_rate, num_iterations):

    losses = []
    for _ in range(num_iterations):
        losses.append(loss_fuction(w, X, y))
        w = gradient_descent_step(w, X, y, learning_rate)
    print(losses)
    return w
```

```python
optimal_w = gradient_descent(x_data, y_data, weight, learning_rate, num_iterations)
print(optimal_w.shape)
optimal_w
```

Знайдіть ці ж параметри за допомогою аналітичного рішення;

```python
def analytical_solution(X, y):
    return np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
analytical_solution = analytical_solution(x_data, y_data)
analytical_solution
```
